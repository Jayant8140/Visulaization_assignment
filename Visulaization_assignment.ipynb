 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Output to the notebook
output_notebook()

# Generate random data
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.randint(5, 20, 50)
# Repeat the colors to match the number of data points
colors = np.tile(['red', 'blue', 'green', 'yellow'], len(x) // 4 + 1)[:len(x)]

# Create a ColumnDataSource
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
source = ColumnDataSource(data)

# Create the plot
p = figure(title="Scatter Plot", x_axis_label="x", y_axis_label="y")

# Add a scatter glyph
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

 4. Create a Bokeh histogram to visualize the distribution of the given data.

 hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import Quad

# Output to the notebook
output_notebook()

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create the plot
p = figure(title="Histogram",
           x_axis_label="Value",
           y_axis_label="Frequency")

# Add a quad glyph to represent the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white")

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper

# Output to the notebook
output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)  # Random data for the heatmap
x = np.linspace(0, 1, 10)  # x-coordinates
y = np.linspace(0, 1, 10)  # y-coordinates

# Create color mapper
colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce",
          "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]  # Color palette
mapper = LinearColorMapper(palette=colors, low=data_heatmap.min(),
                         high=data_heatmap.max())  # Map data to colors

# Create the plot
# Changed plot_width to width and plot_height to height
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1),
           x_axis_location="above", width=300, height=300,
           tools="hover,save", toolbar_location='below',
           tooltips=[('x', '$x'), ('y', '$y'), ('value', '@image')])

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1,
        color_mapper=mapper, level="image")

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
